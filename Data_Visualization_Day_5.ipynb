```python
import matplotlib.pyplot as plt
import numpy  as np
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,label="Sine Wave")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Line Plot")
plt.legend()
plt.show()
```

```python
categories = ['A','B','C','D']
values = [10,25,15,30]
plt.figure(figsize=(6,4))
plt.bar(categories,values,color='purple')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

```python
data = np.random.randn(1000)
plt.figure(figsize=(7,5))
plt.hist(data,bins=30,color='green',edgecolor='black',alpha=0.7)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
data = np.random.randn(1000)
df = pd.DataFrame(data,columns=['Values'])
sns.histplot(df['Values'],bins=30,kde=True,color='Blue')
plt.title("Histogram with KDE")
plt.show()
```

```python
tips = sns.load_dataset("tips")
plt.figure(figsize=(6,4))
sns.boxplot(x=tips['total_bill'])
plt.title("Box Plot of TOtal Bill")
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
numerical_cols = tips.select_dtypes(include=np.number).columns
corr_matrix = tips[numerical_cols].corr()

plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
